```python
import numpy as np
import os
import pandas as pd
```

```python
print(os.listdir('../input'))
```

```python
nRowsRead = None  # specify 'None' if want to read whole file
# podcasts.csv has 121175 rows in reality, but we are only loading/previewing the first 1000 rows
df = pd.read_csv('../input/podcasts.csv', delimiter=',', nrows=nRowsRead)
df.dataframeName = 'podcasts.csv'
nRow, nCol = df.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
df = df[df.language == 'English'][:100000]
df.shape
```

```python
df.drop_duplicates(subset=['itunes_id']).shape
```

```python
dropped = df[df[['title', 'author', 'description']].notnull().all(1)]
dropped = dropped.drop(['uuid', 'language', 'website'], axis=1)
dropped.rename(columns={"itunes_id": "id"}, inplace=True)
dropped.id = dropped.id.astype(str)
print(dropped.shape)
dropped.head(1)
```

```python
with open(f'/kaggle/working/podcasts.jsonl', 'w') as f:
    json = dropped.to_json(orient='records', lines=True)
    f.write(json)
```

```python
with open(f'/kaggle/working/podcasts-1000-samples.jsonl', 'w') as f:
    json = dropped[:1000].to_json(orient='records', lines=True)
    f.write(json)
```
